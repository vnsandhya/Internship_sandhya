```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants')
page
```

```python
sp = BeautifulSoup(page.content)
sp
```

```python
sp_na = sp.find('div', class_="restnt-info cursor")
sp_na
```

```python
sp_na.text
```

```python
lo = sp.find('div',class_="restnt-loc ellipsis")
lo
```

```python
lo.text
```

```python
price = sp.find('span',class_="double-line-ellipsis")
price
```

```python
price.text.replace('₹','')
```

```python
title = []

for i in sp.find_all('div',class_="restnt-loc ellipsis"):
    title.append(i.text)
    
title 
```

```python
loc = []
for i in sp.find_all('div',class_="restnt-loc ellipsis"):
    loc.append(i.text)
    
loc

```

```python
cost = []
for i in sp.find_all('span',class_="double-line-ellipsis"):
    cost.append(i.text.replace('₹',''))
cost
```

```python
image = []
for i in sp.find_all('img',class_="no-img"):
    image.append(i['data-src'])


image
```

```python
print(len(title),len(loc),len(cost),len(image))
```

```python
import pandas as pd
```

```python
table =pd.DataFrame({'Title':title,'location':loc,'Price':cost,'image.url':image })
table
```
